```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data= pd.read_csv("México_sin_nulos.csv")
```

```python
data.info()
```

```python
# Seleccionar columnas numéricas (int y float)
cuantitativas = data.select_dtypes(include=['int64', 'float64'])

# Seleccionar el resto de las columnas (cualitativas)
cualitativas = data.select_dtypes(exclude=['int64', 'float64'])
```

```python
fig = plt.figure(figsize=(15, 8))
cuantitativas.plot(kind="box", vert=False)
plt.title("Valores Atipicos del data frame")
plt.show() #dibujamos el diagrama
```

```python
cuantitativas.info()
```

```python
#Desviacion estandar. 
y=cuantitativas
Limite_Superior= y.mean() +3*y.std()
Limite_Inferior= y.mean() -3*y.std()
print("Limite superior permitidos", Limite_Superior)
print("Limite inferior permitido" , Limite_Inferior)
y
```

```python
data3= cuantitativas[(y<=Limite_Superior) & (y>=Limite_Inferior)]
data3
```

```python
valores_nulos=data3.isnull().sum()
valores_nulos
```

```python
#Reemplzamos los valores atipicos
data_clean=data3.copy()
data_clean=data_clean.fillna(round(data3.mean(),1))
data_clean
```

```python
#Corroboramos valores nulos
valores_nulos=data_clean.isnull().sum()
valores_nulos
```

Aplicando cuartiles


```python
percentile25= y. quantile(0.25)
percentile75= y.quantile(0.75)
iqr=percentile75 - percentile25

Limite_Superior_iqr= percentile75 + 1.5*iqr
Limite_Inferior_iqr= percentile25 - 1.5*iqr
print("Limite superiot permitido", Limite_Superior_iqr)
print("Limite_Inferior permitido", Limite_Inferior_iqr)
```

```python
#Obtenemos datos limpios del Dataframe
data3_iqr= cuantitativas [(y<=Limite_Superior_iqr)&(y>=Limite_Inferior_iqr)]
data3_iqr
```

```python
valores_nulos=data3_iqr.isnull().sum()
valores_nulos
```

```python
data4_iqr=data3_iqr.copy()
data4_iqr=data4_iqr.fillna(round(data3_iqr.mean(),1))
data4_iqr
```

```python
#Unimos el dataframe cuantitativo limpio con el dataframe cualitativo
Datos_Limpios=pd.concat([cualitativas, data4_iqr], axis=1)
Datos_Limpios
```

```python
Datos_Limpios.to_csv("Dstos_Limpios_Mexico.csv")
```
