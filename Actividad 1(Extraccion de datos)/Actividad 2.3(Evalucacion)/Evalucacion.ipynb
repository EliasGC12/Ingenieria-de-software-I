```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= pd.read_excel("listings_50_Datos.xlsx")
```

```python
#Filtro para el inciso a
filtro1=df.head()
filtro1
```

```python
filtro1.to_csv("Entregable_5nombres.csv")
```

```python
#Para el inciso b
filtro2 = df[df["host_since"].dt.year < 2019]
filtro2
```

```python
filtro2.to_csv("Entregable_fechas2019.csv")
```

```python
#Para el insico c
# Rellenar los valores nulos con una cadena vacía
df['host_response_time'] = df['host_response_time'].fillna('')

# Filtrar por la frase "within an hour" sin distinguir entre mayúsculas y minúsculas
filtro3 = df[df["host_response_time"].str.contains("within an hour", case=False)]
filtro3
```

```python
filtro3.to_csv("Entregable_Tiempo_Respuesta.csv")
```

```python
#Para el inciso d
filtro4=df[df["accommodates"] >= 5]
filtro4
```

```python
filtro4.to_csv("Entregable_Accommodates.csv")
```

```python
#Para el inciso e

df['host_response_time'] = df['host_response_time'].fillna('')


filtro5 = df[df["room_type"].str.contains(r"Entire home|Private room", case=False, regex=True)]
filtro5
```

```python
filtro5.to_csv("Entregable_Tipos_de_cuartos.csv")
```

```python
#Para el inciso f
filtro6=df[df["beds"] <= 4]
filtro6
```

```python
filtro6.to_csv("Entregable_Camas.csv")
```

```python
#Para el inciso G
filtro7 = df[(df['bathrooms'] == 1) | 
            (df['host_is_superhost'] != 'f') & 
            (df['instant_bookable'] != 'f')]
filtro7
```

```python
filtro7.to_csv("Entregable_Muchas_Condiciones.csv")
```
